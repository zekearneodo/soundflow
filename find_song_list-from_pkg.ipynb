```python
def try_wrap(func, *args, **kwargs):
    '''
    Wraps a function to return None and log a warning message if it doesn't fall through
    '''
    logger.info('Trying function {}'.format(func.__name__))
    def tried_func(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except:
            logger.warning('Error in function {}')
            return None
    return tried_func


#@try_wrap
def error_test(bird, sess):
    logger.info('Bird {} Sess {}'.format(bird, sess))
    if bird=='z053':
        raise ValueError('Error')
        return None
    return bird, sess

def try_error_test(bird, sess):
    try:
        return error_test(bird, sess)
    except:
        logger.warning('Error in function')
        logger.warning(traceback.print_tb(sys.exc_info()[2]))
        return None
    
```
